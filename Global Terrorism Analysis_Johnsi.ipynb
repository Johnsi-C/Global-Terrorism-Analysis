#### Chart - 1
**Find the number of Terrorist Attacks in each year**

````python
# Chart - 1 visualization code
x_year = GTD['Year'].unique()
y_count_years = GTD['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (25,12))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'rocket_r')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year',fontweight = 'bold')
plt.ylabel('Number of Terrorist Attacks',fontweight = 'bold')
plt.title('Attack_of_Years',fontweight='bold')
plt.show()
````

IN GTD 'year' is a categorical variable. I picked the bar chart to compare the number of terrorist attack in different yaer.

#### Chart - 8
**Analyse attack in Iraq country**

**State wise attack**






````python
# Chart - 8 visualization code
def change_case(text):
    text = text.lower()
    return text[0].upper()+text[1:]
````

````python
Iraq_attacks = GTD[GTD.Country=='Iraq'].reset_index()
Iraq_attacks.State = Iraq_attacks.State.apply(change_case)
Iraq_attacks.head()
````

````python
Iraq_attacks.State.value_counts()
````

````python
plt.subplots(figsize=(10,6))
State_attacks_Iraq = Iraq_attacks.State.value_counts()[:15].reset_index()
State_attacks_Iraq.columns = ['State', 'Total Attacks']
State_attacks_Iraq.drop(1, inplace=True)
sns.barplot(x=State_attacks_Iraq.State, y=State_attacks_Iraq['Total Attacks'], palette='OrRd_r',
            edgecolor=sns.color_palette('dark',10))
plt.xlabel('State',fontweight='bold',fontsize=18)
plt.ylabel('Total Attacks',fontweight='bold',fontsize=14)
plt.xticks(rotation=40)
plt.title('Number Of Total Attacks in Each Iraq State',fontweight = 'bold', fontsize = 18)
plt.show()
````

#### Chart - 10
**Killed Data by attack type**

````python
killData = GTD.loc[:,'Killed']
attackData = GTD.loc[:,'AttackType']
typeKillData = pd.concat([attackData, killData], axis=1)
typeKillFormatData = typeKillData.pivot_table(columns='AttackType', values='Killed', aggfunc='sum')
typeKillFormatData


````

````python
Trans = typeKillFormatData.T
Killed_AttackType = Trans.sort_values(by = 'Killed', ascending = False)
Killed_AttackType
````

````python
Killed_AttackType.plot(kind = 'bar', ylabel = 'Killed', figsize = (12,7))
plt.title('Killed by attack type',fontweight = 'bold', fontsize = 18)
````

##### 1. Why did you pick the specific chart?

First, I used a pivot table to find the count of killed values.

Next, I used a bar chart to show the killed count over attack type
